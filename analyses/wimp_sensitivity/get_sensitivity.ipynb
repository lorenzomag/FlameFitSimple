```python
import os

import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../../helper_classes')
from limit_utils import get_median_sensitivity
```

```python
def scaling_fn(mu, mu_ref):
    return mu / mu_ref * 1e-45
```

```python
masses, bands = get_median_sensitivity(signal_name='WIMP',
                                       likelihood='SI_40t_140ty.pkl',
                                       analysis='.',
                                       directory='../SI_40t_140ty_output',
                                       scaling_fn=scaling_fn,
                                      )
```

```python
fig = plt.figure()
ax = fig.gca()

plt.plot(masses, bands[0], color='black', lw=2.)

ylab = r'WIMP-nucleon $\sigma_{\rm SI}$ [cm$^2$]'
ylim = (5e-50, 1e-46)

ax.set_xlabel(r'WIMP Mass [GeV/c$^2$]')
ax.set_ylabel(ylab)
xmin=8
xmax=10000
ax.set_xlim(xmin,xmax)
ax.set_ylim(ylim)
ax.set_yscale('log')
ax.set_xscale('log')
```
